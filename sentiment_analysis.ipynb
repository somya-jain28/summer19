```
# printing text only
pos=0
neg=0
neu=0
for tweet in tweets:
  analysis=TextBlob(tweet.text)
  #print(analysis.sentiment)
  if analysis.sentiment.polarity > 0 :
    print("positive ")
    pos=pos+1
  elif analysis.sentiment.polarity == 0:
    print("neutral")
    neu=neu+1
  else :
    print("negative")
    neg=neg+1
    
print(pos,neg,neu)    
  
```

```
# plotting graph 
plt.xlabel("sentiments")
plt.ylabel("count")
plt.bar(["positive","negative","neutral"],[pos,neg,neu])
plt.show()
```
